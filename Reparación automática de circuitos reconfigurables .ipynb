```python
# -*- coding: utf-8 -*-
#!/usr/bin/python

import random

from builtins import range

############ OBJETIVO ESPECÍFICO 1 ##############

#Implementación de la estructura de datos:
def creaCircuito(capas, puertas):
    #Con este método creamos un circuito de m capas y n puertas en cada capa
    #En primer lugar definimos el conjunto donde se guardara el circuito
    Circuito = []
    #Definimos los distintos tipos de puertas que puede haber en el circuito
    Tipo = ['OR', 'AND', 'NOT', 'NAND', 'XOR', 'XNOR', 'NOR']
    #Número de puertas que tendrá el circuito que nos servirá de contador
    Total = capas*puertas
    #Contador para definir la capa de cada puerta
    contador = 0
    #Número de capa de la puerta, empieza en 1
    capa = 1
    #Definimos la variable de entrada a las puertas, inicialmente será 0 que nos servirá para establecer la entrada nula 
    conexion1 = 0
    conexion2 = 0
    #Se repetirá para el número total de capas
    for puerta in range(Total):
        #Seleccionamos un tipo de puerta con random
        tipoPuerta = random.choice(Tipo)
        #Si no es una puerta 'not', Asignamos las puertas con las que se 
        #conectarán las entradas llamando al método asignaConexion 
        if tipoPuerta != 'NOT':
            conexionDoble = asignaConexion(puertas, capa)
            conexion1 = conexionDoble[0]
            conexion2 = conexionDoble[1]
        #Asignaremos para una puerta tipo 'not', solo asignamos la conexión2
        if tipoPuerta == 'NOT':
            #Si la puerta se encuentra en una capa>2
            if capa>2:
                #Definimos el número de puerta más bajo elegible
                minimo = (((capa-2)*puertas)-(puertas-1))
                #Definimos el número de puerta más alto elegible
                maximo = ((capa-1)*puertas)
                #Elegimos una puerta random para la conexión
                #Creamos una lista donde se incluyen todos los posibles 
                rango = []
                rango.append(0)
                for a in range(minimo,maximo+1):
                    rango.append(a)
                conexion2 = random.choice(rango)
                conexion1 = 0
            #Si la puerta pertenece a la capa 2
            elif capa==2:
                minimo = (((capa-1)*puertas)-(puertas-1))
                maximo = ((capa-1)*puertas)   
                rango2 = []
                rango2.append(0)
                for a in range(minimo,maximo+1):
                    rango2.append(a)
                conexion2 = random.choice(rango2)
                conexion1 = 0
        #Por último si la puerta pertenece a la capa 1 dejaremos las conexiones (0,0)        
        #Añadimos al circuito la puerta 
        Circuito.append((puerta+1, tipoPuerta, capa, [(conexion1, puerta+1),(conexion2, puerta+1)]))  
        contador += 1
        #Si el contador es igual al número de puertas, es decir, se han creado todas las puertas de una capa
        if contador == puertas:
            #Aumenta el número de capa
            capa+=1
            #Contador se resetea
            contador=0
            
    return Circuito

def asignaConexion(puertas, capa):
    conexion1 = 0
    conexion2 = 0       
    #Si la puerta no es de tipo 'not' le asignaremos las dos conexiones
    if capa>2:
        #Definimos el número de puerta más bajo elegible
        minimo = (((capa-2)*puertas)-(puertas-1))
        #Definimos el número de puerta más alto elegible
        maximo = ((capa-1)*puertas)
        #Elegimos una puerta random para la conexión
        #Creamos una lista donde incluimos las posibles puertas y 0.
        rango = []
        rango.append(0)
        for a in range(minimo,maximo+1):
            rango.append(a)
        conexion1 = random.choice(rango)
        conexion2 = random.choice(rango)
    #Si la puerta pertenece a la capa 2
    elif capa==2:
        minimo = (((capa-1)*puertas)-(puertas-1))
        maximo = ((capa-1)*puertas)
        rango2 = []
        rango2.append(0)
        for a in range(minimo,maximo+1):
            rango2.append(a)  
        conexion1 = random.choice(rango2)
        conexion2 = random.choice(rango2)
    #Denuevo si la puerta pertenece a la capa 1 dejaremos las conexiones en (0,0)         
    return [conexion1,conexion2] 





```

```python
#Método simulador de circuitos con opción de puertas/conexiones defectuosas 
def simularCircuito(Circuito, vector_entrada, conexiones_defectuosas, puertas_defectuosas):
    
    # En primer lugar definimos el vector salida que devolveremos 
    vector_salida = []  
    # Calculamos para cada puerta su salida
    for puerta in range(len(Circuito)):
        # Obtenemos los atributos de la puerta
        
        # Numero
        numPuerta = Circuito[puerta][0]
        # Tipo de puerta
        tipo = Circuito[puerta][1]
        # capa
        capa = Circuito[puerta][2]
        # conexiones
        conexiones = Circuito[puerta][3]
        # entradas, por defecto a cero
        entrada1 = 0
        entrada2 = 0
        
        # En primer lugar si la puerta pertenece al conjunto de puertas marcadas como defectuosas
        # podemos establecer que su salida será cero.
        if numPuerta in puertas_defectuosas:
            # Añadimos al vector salida el correspondiente vector (numPuerta, valor)
            vector_salida.append(0)
            
        # En caso contrario la puerta no estará dañada luego comprobamos el valor de sus entradas
        else:
            # Primera entrada
            # De nuevo comprobamos si la conexion a esta puerta esta dañada en ese caso, el valor
            # de la entrada será cero, por lo tanto no modificamos la entrada.
            
            # En el siguiente caso si la entrada se corresponde con una conexión a 0 quiere decir que o bien está
            # en la capa 1 o no tiene una salida asignada, en cuyo caso la entrada tendrá valor cero. De nuevo 
            # Tampoco modificamos la entrada
            
            # Si la puerta se encuentra en la capa 1 le introducimos el valor del vector entrada correspondiente
            if capa == 1:
                entrada1 = vector_entrada[puerta]
            # Si la conexión no se encuentra en la capa 1, no es defectuosa, y no es 0, le asignamos el valor de la
            # puerta que tenga asociada
            elif capa != 1 and conexiones[0] not in conexiones_defectuosas and conexiones[0][0] != 0:
                # Para la lista de valores de puertas
                # Encontramos el valor de la puerta correspondiente
                val = conexiones[0][0]
                entrada1 = vector_salida[val - 1]
            # Segunda entrada
            # De manera análoga con la segunda entrada
            if capa == 1:
                entrada2 = vector_entrada[puerta]
            # Si la conexión no se encuentra en la capa 1, no es defectuosa, y no es 0, le asignamos el valor de la
            # puerta que tenga asociada
            elif capa != 1 and conexiones[1] not in conexiones_defectuosas and conexiones[1][0] != 0:
                # Para la lista de valores de puertas
                # Encontramos el valor de la puerta correspondiente
                    val2 = conexiones[1][0]
                    entrada2 = vector_salida[val2 - 1]
            # Ahora que ya tenemos definidos los valores de las entradas pasamos 
            # a definir el valor de las salidas de las puertas
            vector_salida.append(calculaSalida(entrada1, entrada2, tipo))
    #Devolvemos las salidas de las n última puertas  
    return vector_salida[-puertas:]              
                        
#Método que calcula la salida según las entradas y las tablas de verdad de las puertas           
def calculaSalida(entrada1, entrada2, tipoPuerta):
    
    salida = 0;
    
    if tipoPuerta == 'OR':
        if entrada1 == 1:
            salida = 1
        elif entrada2 == 1:
            salida = 1
        else:
            salida = 0
    elif tipoPuerta == 'AND':
        if entrada1 == 1 and entrada2 == 1:
            salida = 1
        else:
            salida = 0
    elif tipoPuerta == 'NOT':
        if entrada2 == 0:
            salida = 1
        else:
            salida = 0
    elif tipoPuerta == 'NAND':
        if entrada1 == 1 and entrada2 == 1:
            salida = 0
        else:
            salida = 1
    elif tipoPuerta == 'XOR':
        if entrada1 != entrada2:
            salida = 1
        else:
            salida = 0
    elif tipoPuerta == 'XNOR':
        if entrada1 == entrada2:
            salida = 1
        else:
            salida = 0
    elif tipoPuerta == 'NOR':
        if entrada1 == 0 and entrada2 == 0:
            salida = 1
        else: 
            salida = 0
    return salida



#Método que genera los pares de vectores (entradas, salidas esperadas)
#Con la posibilidad de generarlas con puertas/conexiones dañadas
def coleccionPares(circuito, conexiones_defectuosas, puertas_defectuosas):
    coleccionPares = []
    combinaciones = [0 , 1]
    coleccion = list(product(combinaciones, repeat=puertas))
    for entrada in coleccion:
        coleccionPares.append((entrada, simularCircuito(circuito, entrada, conexiones_defectuosas, puertas_defectuosas,)))
        
    return coleccionPares
    
    
#Método de auto-diagnóstico que compara pares de referencia para saber si un 
#circuito tiene defectos, o para saber los fallos entre pares de referencia
#de distintos circuitos
def auto_diagnostico(circuito, circuito2, conexiones_defectuosas, puertas_defectuosas):
    #Creamos la coleccion de pares de referencia sin fallos
    coleccionCorrecta = coleccionPares(circuito, [],[])
    fallos = 0
    
    for n in range(len(coleccionCorrecta)):
        #Para cada vector de entrada simulamos el vector salida aplicando los defectos
        coleccionDefectuosa = simularCircuito(circuito2, coleccionCorrecta[n][0], conexiones_defectuosas, puertas_defectuosas)
        #Para cada elemento del vector salida
        for com in range(len(coleccionDefectuosa)):
            #Comparamos la salida esperada con la salida obtenida con defectos
            if coleccionDefectuosa[com]!=coleccionCorrecta[n][1][com]:
           
                #Si no coinciden incrementamos los fallos
                fallos+=1
    
    return fallos
    
    
```

```python

#Pasamos ahora a generar la función gen

#Información consultada sobre los genes.
#https://www.researchgate.net/publication/255599933_Logic_Circuits_Synthesis_Through_Genetic_Algorithms

#Para cada cromosoma, en este caso puertas, tendremos 3 genes que vendrán a ser: tipo de puerta, entrada1 y entrada2
#Para la generación de los genes entonces, en primer para las puertas, escogeremos un numero random entre el número
#que después se asginará. Por otro lado para las puertas  


from functools import wraps

#Creamos una función decorator para que nos devuelva el número de veces
#que el método creaGenes ha sido llamado, en este caso por register
def counter(func):
    @wraps(func)
    def tmp(*args, **kwargs):
        tmp.count += 1
        return func(*args, **kwargs)
    tmp.count = 0
    return tmp


#método auxiliar que devuelve la capa de la puerta
def numeroCapa(contador):
    capa = capas
    for n in range(1,capas+1):
        if contador >= ((n*(puertas*3))-((puertas*3)-1)) and contador <= (n*(puertas*3)):
            capa = n
    return capa

@counter
def creaGenes():
    #Para poder saber que tipo de gen se está creando en cada comento (puerta, entrada)
    #Vamos a implementar un método (decorator) que nos de el número de veces que
    #se llama al método, ya que no se puede introducir un contador en el método.
    #Con este contador definimos el número de capa y si tipo de gen.
    if creaGenes.count == (3*puertas*capas):
        creaGenes.count = 0
    contador = creaGenes.count
    gen = 0
    capa = numeroCapa(contador)
    #Si el gen que se está tratando es un tipo de puerta entonces
    #le asignamos un número random del conjunto de tipos de puertas
    if (contador-1) % 3 == 0:
        Tipo = [1,2,3,4,5,6,7]
        gen = random.choice(Tipo)
    #Si son genes relacionados con entradas se les asignarán los correspondientes
    #valores random dentro de su conjunto permitido
    elif (contador-1) % 3 != 0 and capa == 1:
        #Si la capa es la 1
        gen = 0
    elif (contador-1) % 3 != 0 and capa == 2:
        #Si es de la capa 2
        minimo = (((capa-1)*puertas)-(puertas-1))
        maximo = ((capa-1)*puertas)
        rango2 = []
        rango2.append(0)
        for a in range(minimo,maximo+1):
            rango2.append(a)
        gen = random.choice(rango2)        
    elif (contador-1) % 3 != 0 and capa > 2: 
        #Si es de una capa superior a la 2
        minimo = (((capa-2)*puertas)-(puertas-1))
        maximo = ((capa-1)*puertas)
        rango = []
        rango.append(0)
        for a in range(minimo,maximo+1):
            rango.append(a)
        gen = random.choice(rango)        
    return gen
    
#Ahora si podemos registrar en la toolbox la función creaGenes()
caja_de_herramientas.register('gen', creaGenes)

#Registramos ahora en la caja de herramientas la funcion individuo que devuelve
#un individuo aleatorio, definiendo n como el número de genes del que se compone
#el individuo.
caja_de_herramientas.register('individuo', tools.initRepeat, container = creator.Individuo,
                             func = caja_de_herramientas.gen, n= capas*puertas*3)

#Ahora con una lista de individuos podemos crear también una población
#Definiendo n como el número de individuos de la población.
caja_de_herramientas.register('poblacion', tools.initRepeat, container = list,
                             func = caja_de_herramientas.individuo, n = 20)

#Ahora vamos a registrar la función que evalua el fenotipo de cada individuo
def fenotipo(individuo):
    #Devolveremos un fenotipo que este caso debe seguir la estructura de un circuito
    fenotipo = []
    contadorCapa = 0
    #Número de capa de la puerta, empieza en 1
    capa = 1 
    #número de puerta
    puerta = 1
    #vamos a parsear los genes del individuo para formar el fenotipo
    for gen in range(0, len(individuo), 3):
        #Parseamos el número de tipo de puerta al nombre del tipo
        tipoPuerta = tipo_puerta(individuo[gen])
        
        fenotipo.append((puerta, tipoPuerta, capa, [(individuo[gen+1], puerta),(individuo[gen+2], puerta)]))
        puerta+=1
        contadorCapa += 1
        #Si el contador es igual al número de puertas, es decir, se han creado todas las puertas de una capa
    
        if contadorCapa == puertas:
            #Aumenta el número de capa
            capa+=1
            #Contador se resetea
            contadorCapa=0
    return fenotipo

#método auxiliar para obtener el tipo de puerta según el número de gen
def tipo_puerta(numero):
    puerta = 'NOR'
    if numero == 1:
        puerta = 'OR'
    elif numero == 2:
        puerta = 'AND'
    elif numero == 3:
        puerta = 'NOT'
    elif numero == 4:
        puerta = 'NAND'
    elif numero == 5:
        puerta = 'XOR'
    elif numero == 6:
        puerta == 'XNOR'
    return puerta
    
    
    
def evaluar_individuo(individuo):
    
    #Fenotipo para el individuo
    circuito = fenotipo(individuo)
    #Ahora tenemos en cuenta el circuito que en primer lugar tomamos
    #como circuito solución, suponiendo que este resulta dañado
    #y lo comparamos con el fenotipo del individuo. Utilizamos la
    #función de auto-diagnóstico que hemos implementado antes
    return [auto_diagnostico(CircuitoSolucion, circuito, [], [])]
    
    
#Incluimos en la toolbox la función evaluate    
caja_de_herramientas.register('evaluate', evaluar_individuo)



```

```python
#Ahora que tenemos todas las funciones registradas, pasamos a 
#registrar los operadores que utilizaremos en el algorítmo 
#genético, operadores de cruce y mutación. Vamos a seleccionar 
#solo operadores factibles para nuestro dominio, para luego no 
#tener que penalizar las funciones fitness.

#Teniendo en cuenta que estamos usando genotipos con representación 
#de números enteros podemos aplicar los operadores de cruzamiento de 
#recombinación en 1 punto, 2 puntos o uniforme. Ya que se intercambiarán
#genes en las mismas posiciones sin crear problemas de coherencia.
#Elegimos la recombinación en dos puntos para que haya la posibilidad de
#recombinar todas las partes del circuito.

# EJECUTAR SOLO EL MÉTODO DE CRUZAMIENTO QUE SE QUIERA USAR

caja_de_herramientas.register('mate', tools.cxTwoPoint)
```

```python
#De igual manera incluimos el operador de recombinación uniforme, estableceremos
#la probabilidad de intercambio en un 30%

caja_de_herramientas.register('mate', tools.cxUniform,  indpb=0.3)
```

```python
#Por otro lado tenemos la mutación que para nuestro problema está claro
#que el operador más coherente es el de mutación uniforme, ya que se puede
#especificar un intervalo, en nuestro caso según la capa en la que esté la
#puerta especificaremos un intevervalo u otro.


#método auxiliar para crear los márgenes para la mutación
def creaMargenes():
    lowerBound = []
    upperBound = []
    for n in range(puertas*capas*3):
        capa = numeroCapa(n+1)
        if n % 3 == 0:
            lowerBound.append(1)
            upperBound.append(7)
        elif n % 3 != 0 and capa == 1:
            lowerBound.append(0)
            upperBound.append(0)
        elif n % 3 != 0 and capa == 2:
            minimo = (((capa-1)*puertas)-(puertas-1))
            maximo = ((capa-1)*puertas)
            #Tenemos que dejar de incluir 0 como posibilidad de entrada a las puertas
            #Ya que las secuencias de márgenes introducirían valores no factibles en 
            #los genotipos.
            lowerBound.append(minimo)
            upperBound.append(maximo)        
        elif n % 3 != 0 and capa > 2: 
            minimo = (((capa-2)*puertas)-(puertas-1))
            maximo = ((capa-1)*puertas)
            lowerBound.append(minimo)
            upperBound.append(maximo)
    return [lowerBound,upperBound]

bounds = creaMargenes()

#Lo introducimos en la toolbox
caja_de_herramientas.register('mutate', tools.mutUniformInt,low = bounds[0],up = bounds[1], indpb=0.1)
```
